**Data Manipulation**

**4.0 Joining Data**
* product.id from orderlines.xlsx are the same as bike.id from bikes.xlsx
* Therefore they are combined (both dataset share product.id)

```r
left_join(orderlines_tbl, bikes_tbl, by = c("product.id" = "bike.id"))
```

* Combining product.id from orderlines.xlsx and bike.id from bikes.xlsx
* Because the bikes are the product
-----------------------------------
* Combining customer.id from orderlines.xlsx and bikeshop.id from bikeshops.xlsx
* Because the bikeshops (middle man) are buying and selling to customers, therefore they are the customer of the supplier (Seller)
* The individual people who buys, have data recorded by each shop, not important to this case. 

```r
bike_orderlines_joined_tbl = orderlines_tbl %>%
    left_join(bikes_tbl, by = c("product.id" = "bike.id")) %>% 
    left_join(bikeshops_tbl, by = c("customer.id" = "bikeshop.id"))

head(bike_orderlines_joined_tbl, n = 20)
```

```r
bike_orderlines_joined_tbl %>% glimpse()
```

**5.0 Wrangling Data**
* Data manipulation and cleaning
* unique() will isolate only unique values, no same values collected

```r
bike_orderlines_joined_tbl %>% 
    dplyr::select(category) %>% 
    filter(str_detect(category, "Mountain")) %>%
    unique()
```

* Analyse the product sales by product family (category)
* Add total.price variable (then name changed to total_price with set_names())
* Get rid of ...1 and gender variable
* Cleaning up names that has .id 
* Combine order.id variable from bike_orderlines_joined_tbl into bike_orderlines_wrangled_tbl
* Isolating order.id, every "order" variable, every "model" variable, every "category" variable, price, quantity, total.price, and everything()
* (everything() part will make the isolation redundant)
* rename name variable to bikeshop 
* names() select every name with '.' Then change (replace) every '.' with a '_' (must use "\ \")

```r
bike_orderlines_wrangled_tbl = bike_orderlines_joined_tbl %>%
    separate(col = category,
            into = c("category.1", "category.2", "category.3"),
            sep = " - ") %>%
    mutate(total.price = price * quantity) %>%
    dplyr::select(-...1, -gender) %>%
    dplyr::select(-ends_with(".id")) %>%
    bind_cols(bike_orderlines_joined_tbl %>% dplyr::select(order.id)) %>%
    dplyr::select(order.id, contains("order"), contains("model"), contains("category"),
          price, quantity, total.price, everything()) %>%
    rename(bikeshop = name) %>%
    set_names(names(.) %>% str_replace_all("\\.", "_"))
```

```r
head(bike_orderlines_wrangled_tbl, n = 20)
```

**Insights and Savings**

**6.0 Business Insights**
* Creating business insights
* Analyse Sales by year (revenue)
* Analyse Sales by category

```r
library(lubridate)
```

* Select order date and total price
* select only the year from order date
* Add y new variable year_1
* Group rows by year and insert into year_1
* group_by() %>% summarize() will put together values from summarize to the corresponding group in group_by, automatically mutate()
* add a new variable sales_text and scaled to match a currency format

```r
sales_by_year_tbl = bike_orderlines_wrangled_tbl %>% 
    dplyr::select(order_date, total_price) %>%
    mutate(year_1 = year(order_date)) %>%
    group_by(year_1) %>% 
    summarize(sales = sum(total_price)) %>% 
    mutate(sales_text = scales::dollar(sales,
                                      big.mark = ".",
                                      decimal.mark = ",",
                                      prefix = "",
                                      suffix = "€"))

sales_by_year_tbl
    
```

* Plot data

```r
# Step 2 - Visualize
sales_by_year_tbl %>%

    # Setup canvas with the columns year (x-axis) and sales (y-axis)
    ggplot(aes(x = year_1, y = sales)) +

    # Geometries
    geom_col(fill = "#2DC6D6") + # Use geom_col for a bar plot
    geom_label(aes(label = sales_text)) + # Adding labels to the bars
    geom_smooth(method = "lm", se = FALSE) + # Adding a trendline

    # Formatting
    # scale_y_continuous(labels = scales::dollar) + # Change the y-axis. 
    # Again, we have to adjust it for euro values
    scale_y_continuous(labels = scales::dollar_format(big.mark = ".", 
                                                      decimal.mark = ",", 
                                                      prefix = "", 
                                                      suffix = " €")) +
    labs(
        title    = "Revenue by year",
        subtitle = "Upward Trend",
        x = "", # Override defaults for x and y
        y = "Revenue"
    )
```

* Sales by category

```r
sales_by_year_cat_1_tbl <- bike_orderlines_wrangled_tbl %>%
 
  # Select columns and add a year
  dplyr::select(order_date, total_price, category_1) %>%
  mutate(year = year(order_date)) %>%

  # Group by and summarize year and main catgegory
  group_by(year, category_1) %>%
  summarise(sales = sum(total_price)) %>%
  ungroup() %>%

  # Format $ Text
  mutate(sales_text = scales::dollar(sales, big.mark = ".", 
                                     decimal.mark = ",", 
                                     prefix = "", 
                                     suffix = " €"))

sales_by_year_cat_1_tbl  
```

* Sales by category visualize

```r
sales_by_year_cat_1_tbl %>%

  # Set up x, y, fill
  ggplot(aes(x = year, y = sales, fill = category_1)) +

  # Geometries
  geom_col() + # Run up to here to get a stacked bar plot

  # Facet
  facet_wrap(~ category_1) +

  # Formatting
  scale_y_continuous(labels = scales::dollar_format(big.mark = ".", 
                                                    decimal.mark = ",", 
                                                    prefix = "", 
                                                    suffix = " €")) +
  labs(
    title = "Revenue by year and main category",
    subtitle = "Each product category has an upward trend",
    fill = "Main category" # Changes the legend name
  )
```

**Analyse sales by location (State) with bar plot**

* state and city should be split
* Which state has the highest revenue
* Replace bike_order_wrangled_tbl with wrangled object with column state and city 

```r
location_splitted = separate(
    bike_orderlines_wrangled_tbl,
    col = location,
    into = c("city", "state"),
    sep = ",")

location_splitted
```

```r
sales_by_state_tbl = location_splitted %>%
    dplyr::select(state, total_price) %>%
    mutate(state_1 = state) %>%
    group_by(state_1) %>% 
    summarize(sales = sum(total_price)) %>%
    mutate(sales_text = scales::dollar(sales,
                                      big.mark = ".",
                                      decimal.mark = ",",
                                      prefix = "",
                                      suffix = "€"))

sales_by_state_tbl
```

```r
dev.new(width=7, height=6, unit="in")
plot(1:20)
dev.new(width = 750, height = 530, unit = "px")
plot(1:15)
```

```r
sales_by_state_tbl %>%
    ggplot(aes(x = state_1, y = sales)) +
    geom_col(fill = "#2DC6D6") + 
    geom_label(aes(label = sales_text)) +
    scale_y_continuous(labels = scales::dollar_format(big.mark = ".",
                                                     decimal.mark = ",",
                                                     prefix = "",
                                                     suffix = "€")) +
    theme(axis.text.x = element_text(angle = 45, hjust = 1)) +

    labs(
        title = "Revenue by State",
        subtitle = "",
        x = "state",
        y = "Revenue"
    )
```
